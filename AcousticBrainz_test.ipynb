```python
import pandas as pd

df_charts = pd.read_csv("charts.csv")

df_charts.head()
```

```python
df_charts['date'] = pd.to_datetime(df_charts['date'])

df_2010 = df_charts[
    (df_charts['date'].dt.year == 2010) &
    (df_charts['rank'] <= 10)
]

print("Total Top 10 entries in 2010:", len(df_2010))
```

```python
sample_songs = df_2010.sample(20, random_state=42)

sample_songs[['song', 'artist']]
```

```python
import requests
import time

def get_mbid(title, artist):
    query = f'recording:"{title}" AND artist:"{artist}"'
    url = "https://musicbrainz.org/ws/2/recording/"
    params = {
        "query": query,
        "fmt": "json"
    }
    
    r = requests.get(url, params=params)
    if r.status_code != 200:
        return None
    
    data = r.json()
    
    if data.get("recordings"):
        return data["recordings"][0]["id"]
    
    return None

def check_acousticbrainz(mbid):
    url = f"https://acousticbrainz.org/{mbid}/low-level"
    r = requests.get(url)
    return r.status_code == 200
```

```python
found = 0
total = len(sample_songs)

for _, row in sample_songs.iterrows():
    title = row['song']   # adjust if needed
    artist = row['artist']
    
    mbid = get_mbid(title, artist)
    
    if mbid and check_acousticbrainz(mbid):
        found += 1
    
    time.sleep(1)

print("Found:", found)
print("Coverage %:", found / total * 100)
```
